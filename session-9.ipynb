```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
random_numbers = np.random.uniform(0, 1, 1000)
```

```python
plt.hist(random_numbers, bins=100, color="green", edgecolor="purple")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.grid(True)
plt.savefig("session-9.pdf",bbox_inches="tight",facecolor="white")
```
