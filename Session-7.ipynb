```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
def exp_function(x):
    return np.exp(x)
```

```python
y = exp_function(x)
```

```python
plt.figure(figsize=(7,7))
plt.plot(x, y, color='red', label='exp(x)')
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.title('Plot for Session 7')
plt.legend()
plt.grid(True)
plt.savefig("session-7.pdf",bbox_inches="tight",facecolor="white")
```
