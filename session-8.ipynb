```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1, 2, 1)
plt.plot(x, sin_function(x), color='blue')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin(x)')

plt.subplot(1, 2, 2)
plt.plot(x, cos_function(x), color='green')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('cos(x)')

plt.savefig("Session-8.pdf", bbox_inches="tight", facecolor="white")
```
